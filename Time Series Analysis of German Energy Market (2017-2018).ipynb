```python
# Visualizing Time Series Plot of Actual total Load 
from pandas.plotting import register_matplotlib_converters
df = pd.read_csv('OPSD_60min_interval_energy_data.csv')

DE_total_load = list(df.loc[:,'DE_load_actual_entsoe_transparency'])


plt.style.use('seaborn')

dateslist = []
for i in range(len(df.iloc[:,0])):
	dateslist.append(parser.parse(df.iloc[i,0]))
	
plt.plot_date(dateslist, DE_total_load,linestyle='solid',marker='')


plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%d/%b/%Y')

plt.gca().xaxis.set_major_formatter(date_format)

plt.title('Time Series plot of Actual Total Load in Germany between 2017 and 2019')
plt.ylabel("Total Actual Load MW")
plt.xlabel("Time between 2017-2019")
plt.tight_layout()


plt.plot()
plt.show()
```

```python
# DECOMPOSITION PLOTS for Solar and Wind generation
df = pd.read_csv("tseries.csv",parse_dates=['utc_timestamp'],
                     index_col='utc_timestamp')
wg=df[["DE_wind_onshore_generation_actual"]]
sg=df[["DE_solar_generation_actual"]]

#Aggregating per months 
wind_gen_agg=wg.resample('M').mean()

sol_gen_agg=sg.resample('M').mean()

print(wind_gen_agg.head()) # Printing agg values per month for wind generation

print(sol_gen_agg.head()) # Agg. values per month for Solar energy generation

#Plotting decomposition

from statsmodels.tsa.seasonal import seasonal_decompose
series = sol_gen_agg
result = seasonal_decompose(series, model='additive')
result.plot()

from statsmodels.tsa.seasonal import seasonal_decompose
series = wind_gen_agg
result = seasonal_decompose(series, model='additive')
result.plot()

flights = sns.load_dataset("tseries.csv")
flights = flights.pivot("month", "year", "DE_solar_generation_actual")
ax = sns.heatmap(flights)
```

```python
#Heatmap for Solar generation
df = pd.read_csv("OPSD_60min_interval_energy_data.csv",parse_dates=['utc_timestamp'],
                     index_col='utc_timestamp')
date_rng = pd.date_range(start='1/1/2017', end='2018-12-31 23:00:00', freq='H')

mod  = pd.DataFrame(list(df['DE_solar_generation_actual']), columns = ['DE_solar_generation_actual'], index=date_rng)
mod['timeStamp'] = date_rng
mod['Months'] = mod['timeStamp'].apply(lambda x: x.strftime('%Y %m/%d'))
mod['Hour'] = mod['timeStamp'].apply(lambda x: x.strftime('%H'))

second_mod =pd.pivot_table(mod, values='DE_solar_generation_actual', index=['Hour'] , columns=['Months'])

ax = sns.heatmap(second_mod,cmap="hot")

```

```python
#Decomposition Plot for Day Ahead Prices

df = pd.read_csv("OPSD_60min_interval_energy_data.csv",parse_dates=['utc_timestamp'],
                     index_col='utc_timestamp')
dfw=df[["DE_price_day_ahead"]]

day_ahead_agg=dfw.resample('M').mean()


print(day_ahead_agg.head())


from statsmodels.tsa.seasonal import seasonal_decompose

series = day_ahead_agg
result = seasonal_decompose(series, model='additive')
result.plot()

flights = sns.load_dataset("OPSD_60min_interval_energy_data.csv")
flights = flights.pivot("month", "year", "DE_price_day_ahead")
ax = sns.heatmap(flights)
```

```python
#Heatmap for day Ahead Prices 


date_rng = pd.date_range(start='1/1/2017', end='2018-12-31 23:00:00', freq='H')

mod  = pd.DataFrame(list(df['DE_price_day_ahead']), columns = ['DE_price_day_ahead'], index=date_rng)
mod['timeStamp'] = date_rng
mod['Months'] = mod['timeStamp'].apply(lambda x: x.strftime('%Y %m/%d'))
mod['Hour'] = mod['timeStamp'].apply(lambda x: x.strftime('%H'))

second_mod =pd.pivot_table(mod, values='DE_price_day_ahead', index=['Hour'] , columns=['Months'])

ax = sns.heatmap(second_mod,cmap="YlGnBu")
```

```python
# Heatmap for Onshore Wind energy Generation 

mod  = pd.DataFrame(list(df['DE_wind_onshore_generation_actual']), columns = ['DE_wind_onshore_generation_actual'], index=date_rng)
mod['timeStamp'] = date_rng
mod['Months'] = mod['timeStamp'].apply(lambda x: x.strftime('%Y %m/%d'))
mod['Hour'] = mod['timeStamp'].apply(lambda x: x.strftime('%H'))

second_mod =pd.pivot_table(mod, values='DE_wind_onshore_generation_actual', index=['Hour'] , columns=['Months'])

ax = sns.heatmap(second_mod,cmap="YlGnBu")
```
